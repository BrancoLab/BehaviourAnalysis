```python
# Define functions

def sigmoid(x, L=1, x0=.5, k=10, b=0):
    """
    L -> shirnks the function on the Y axis. 
    x0 -> x shift. 
    k  -> slope. the smaller the flatter. Vals > 5 are good
    b -> y shift. Shifts thw whole curve up and donw
    """ 
    return L / (1 + np.exp(-k*(x-x0)))+b

def sigma1(x):
    return sigmoid(x, L=-.5, x0=.5, k=10, b=1)

def sigma2(x):
    return sigmoid(x, L=.5, x0=.5, k=-10, b=0)

def fPsi(x, y, k=10):
    n = 1
    d = 1 + np.exp(-k/y * (x-(1-y))) + 0
    return 1 - n/d
```

```python
# %matplotlib notebook
%matplotlib inline

# ? Plot Psi as a 2d image
f, axarr = create_figure(subplots=True, ncols=2)
ax, ax2 = axarr[0], axarr[1]

x = np.arange(0, 1, .01)
y = np.arange(0, 1, .01)

l = len(x)-1
Z = np.zeros((len(x), len(x)))
PSI = np.zeros_like(Z)

for i, xx in enumerate(x):
    for ii, yy in enumerate(y):
        # Z[l-i, ii] = fPsi(xx, yy, k=10)
        psi = yy/xx
        PSI[ii, i] = psi
        Z[ii, i] = sigmoid(psi, L=1, x0=1, k=20, b=0)


surf = ax.imshow(Z, extent=[0, 1, 0, 1], cmap=cm.coolwarm, origin="lower", aspect="equal", vmin=0, vmax=1)
f.colorbar(surf)

surf = ax2.imshow(PSI, extent=[0, 1, 0, 1], cmap=cm.coolwarm, origin="lower", aspect="equal")
f.colorbar(surf)

_x = [.2, .5, .8]

for xx in _x:
    ax.axvline(xx, color="k", alpha=.5)
    ax.axhline(xx, color="k", alpha=.5)


ax.set(title="$\Psi(\psi)$", xlabel="X", ylabel="Y")
ax2.set(title="$\psi$", xlabel="X", ylabel="Y")



print("Max - {}, min: {}. p[0, 0] = {}, p[1, 1]={}".format(
    round(np.max(Z), 3), round(np.min(Z), 3), round(Z[0, 0], 3), round(Z[-1, -1]), 3))
```

```python
%matplotlib inline


# ? Compute and pLot Psi as a surface
fig = plt.figure(figsize=(16, 16))
ax = fig.gca(projection='3d')


x = np.arange(0, 1, .01)
y = np.arange(0, 1, .01)
X, Y = np.meshgrid(x, y)

Z = np.zeros_like(X)
l = len(x)-1

for i, xx in enumerate(x):
    for ii, yy in enumerate(y):
        Z[ii, l-i] = fPsi(xx, yy)


# Plot the surface.
surf = ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, lw=2,  shade=True,
                       antialiased=False, vmin=0, vmax=1)
# surf = ax.contourf(X, Y, Z, cmap=cm.coolwarm, lw=2,  shade=True,
                    #    antialiased=False)


# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)
ax.set(xlabel="X", ylabel="Y", zlabel="Psi")
ax.view_init(20, 135)

print("Max - {}, min: {}. [0, 0] = {}, [1, 1]={}".format(
    round(np.max(Z), 3), round(np.min(Z), 3), round(Z[0, 0], 3), round(Z[-1, -1]), 3))
```
